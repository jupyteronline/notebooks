```python
import pandas as pd
salary=pd.read_csv('http://wolfpack.hnu.ac.kr/Big_Data/data/MLB1871_2018/Salaries.csv')
bat=pd.read_csv('http://203.247.53.31/Big_Data/data/MLB1871_2018/Batting.csv')
df_sal=salary[(salary.yearID==2016) & (salary.salary>0)]; df_bat=bat[bat.yearID==2016]
df=pd.merge(df_sal,df_bat,on='playerID')
df.sort_values(['playerID','stint'], ascending=False,inplace = True) 
df['move']=df['playerID'].duplicated() 
df0=df[(df['move']==False)&(df.stint==1)]
df0.drop(['yearID_x','yearID_y','teamID_y','lgID_y','stint','move'],axis=1,inplace=True)
df0.set_index('playerID',inplace=True)
df0.dropna(inplace=True)
df0.head(3)
```

선수연봉 중위값 기준 상, 하 연봉 집단 나누기

```python
df0['group']=['up' if x>=df0.salary.quantile(0.5) else 'dn' for x in df0['salary']]
df0.head(3)
```

Fisher Quardratic DA - Fisher 2 차 판별 분석

선형판별분석법(linear discriminant analysis, LDA)과 이차판별분석법(quadratic discriminant analysis, QDA)는 대표적인 확률론적 생성모형(generative model)이다. 가능도 즉, y의 클래스값에 따른 x의 분포에 대한 정보를 먼저 알아낸 후, 베이즈 정리를 사용하여 주어진 x에 대한 y의 확률분포를 찾아낸다.

```python
import numpy as np
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
X=df0.iloc[:,3:20].values
y=df0.iloc[:,20].values
qda=QDA()
X_qda=qda.fit(X, y)
```

```python
qda.predict_proba(X)[0:3] #Estimate probability
```

```python
y_pred=pd.DataFrame(qda.predict(X))
y_pred.columns=['group_qda']
y_pred.set_index(df0.index,inplace=True)
y_pred.head(3)
```

```python
df_qda=pd.concat([df0,y_pred],axis=1)
```

```python
df_qda
```

```python
df_qda['DA_result']=df_qda.group+'-'+df_qda.group_qda
df_qda
```

### pd.crosstab : 
    
요인(factors)별로 교차분석(cross tabulations) 해서, 행, 열 요인 기준 별로 빈도를 세어서 도수분포표(frequency table), 교차표(contingency table) 를 만들어준다

```python
qda_table=pd.crosstab(df_qda.group,df_qda.group_qda)
qda_table
```

```python
qda_table.apply(lambda r: r/r.sum(), axis=1) #correct ratio 
```

```python
df00=df0.iloc[:,3:20]
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(df00) #scale standardization
from sklearn.decomposition import PCA
pca=PCA(n_components=2) #80% rule =0.8
df_pca=pca.fit_transform(X) #PC variables
pca_df=pd.DataFrame(df_pca,columns=['AB**','SO**']) 
pca_df.set_index(df00.index,inplace=True)
pca_df.head(3)
```

```python
df_fin=pd.concat([df_qda,pca_df],axis=1)
df_fin=df_fin.loc[:,['DA_result','AB**','SO**']]
df_fin.head()
```

```python
import seaborn as sns
sns.pairplot(df_fin, hue='DA_result')
```

Fisher Linear DA

```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
X=df0.iloc[:,3:20].values
y=df0.iloc[:,20].values
lda=LDA()
X_lda=lda.fit(X, y)
y_pred2=pd.DataFrame(lda.predict(X))
y_pred2.columns=['group_lda']
y_pred2.set_index(df0.index,inplace=True)
df_lda=pd.concat([df0,y_pred2],axis=1)
df_lda['DA_result']=df_lda.group+'-'+df_lda.group_lda
lda_table=pd.crosstab(df_lda.group,df_lda.group_lda)
lda_table.apply(lambda r: r/r.sum(), axis=1)
```

로지스틱 판별분석

```python
from sklearn.linear_model import LogisticRegression
X=df0.iloc[:,3:20].values
y=df0.iloc[:,20].values
logit_fit=LogisticRegression(fit_intercept=True).fit(X, y)
```

```python
y_pred=logit_fit.predict(X)
print('Accuracy : {:.2f}'.format(logit_fit.score(X, y)))
```

```python
y_pred[0:3]
```

```python
logit_fit.predict_proba(X)[0:3]
```

```python
#분류성능평가지표 :
#  - Precision(정밀도) : True라고 분류한 것 중에서 실제 True인 것의 비율
#  - Recall(재현율) : 실제 True인 것 중에서 모델이 True라고 예측한 경우
#  - Accuracy(정확도) : True를 True라고, False를 False라고 예측한 경우
#  - F1 score - Precision과 Recall의 조화평균
#  - Support(지지도) :두 항목 X와 Y의 지지도는 전체 거래 건수 중에서 항목집합 X와 Y를 모두 포함하는 거래 건수의 비율.
#                     지지도는 좋은 규칙(빈도가 많은, 구성비가 높은)을 찾거나, 불필요한 연산을 줄일 때(pruning, 가지치기)의 기준으로 사용


from sklearn.metrics import classification_report
print(classification_report(y, y_pred))
```

```python
table=pd.crosstab(df0.group,y_pred)
table
```

```python
table.apply(lambda r: r/r.sum(), axis=1)
```
