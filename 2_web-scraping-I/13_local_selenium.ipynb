```python
import time
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
```

```python
# 1. 네이버로 이동
```

```python
browser.get("http://naver.com")
```

```python
# 2. 로그인 버튼으로 이동
elem = browser.find_element_by_class_name("link_login")
```

```python
elem.click()
```

```python
# 3. id, pw 입력
browser.find_element_by_id("id").send_keys("aaaaaaaaaa")
browser.find_element_by_id("pw").send_keys("bbbbbbbbbb")
```

```python
# 4. 로그인 버튼 클릭
browser.find_element_by_id("log.login").click()
```

```python
# 5. (로긴 실패 시) ID를 새로 입력
browser.find_element_by_id("id").clear()
browser.find_element_by_id("id").send_keys("aaaaaaaaaa")
```

```python
browser.find_element_by_id("pw").clear()
browser.find_element_by_id("pw").send_keys("bbbbbbbbbb")
```

```python
# 6. html 정보 출력
print(browser.page_source)
```

```python
# 7. 브라우저 종료
browser.quit()
```
